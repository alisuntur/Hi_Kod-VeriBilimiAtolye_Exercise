```python
import pandas as pd
import random

# Veri boyutunu belirle
data_size = 100  # Ã–rnek iÃ§in 100 veri noktasÄ±

# Rastgele veri oluÅŸturma
data = {
    "in_spotify_playlists": [random.randint(0, 10) for _ in range(data_size)],
    "in_apple_playlists": [random.randint(0, 10) for _ in range(data_size)],
    "key": [random.choice(["C#","G","G#","F","B","D","A","F#","E","A#","D#"]) for _ in range(data_size)]
}

# Veriyi DataFrame'e dÃ¶nÃ¼ÅŸtÃ¼rme
df = pd.DataFrame(data)

# Åžimdi veri Ã§erÃ§evesini kullanarak benzer bir kod oluÅŸturabilirsiniz
# Ã–rnek bir iÅŸlem:
grouped_data = df.groupby(["key"])[["in_spotify_playlists", "in_apple_playlists"]].mean().reset_index()

# OluÅŸturulan veriyi gÃ¶sterme
print(grouped_data.head())


```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Ã‡ubuk grafik oluÅŸturma
plt.figure(figsize=(10,12))
sns.barplot(x="in_spotify_playlists", y="in_apple_playlists", hue="key", data=data)
plt.title("Spotify ve Apple MÃ¼zik Listeleri KarÅŸÄ±laÅŸtÄ±rmasÄ±")
plt.xlabel("Spotify Playlists")
plt.ylabel("Apple Music Playlists")

# Grafik gÃ¶sterme
plt.show()
```
